# Math

- `-` minus
- `*` multiplication
- `/` division
- `**` exponent
- `%` modulo

# If Statements
You can use `if`, `elif` and `else` to check things in Python.

```python
answer = input("Do you want to hear a joke? ")

if answer.lower() == "yes":
    print("I'm against picketing, but I don't know how to show it.")
    # Mitch Hedberg (RIP)
elif answer.lower() == "no":
    print("Fine.")
else:
    print("I don't understand.")
```

# Lists
In Python, lists are a way of grouping together similar things (usually)

```python
the_count = [1, 2, 3, 4, 5]
stocks = ["FB", "AAPL", "NFLX", "GOOGL"]
random_things = [55, 1/2, "Puppies", stocks]
```

```python
people = []
people.append("Mattan")
people.append("Jennifer")
people.append("Helena")
people.remove("Jennifer")
print(people)
```

```python
cities = "New York, San Francisco, London".split(", ")
print(cities)
```

```python
last_city = cities[-1]
print(last_city)
```

```python
us_cities = cities[0:2]
print(us_cities)
```

# Loops
Loops let you run the same code for every item in list

```python
numbers = [1, 2, 3]
for number in numbers:
    print(number)
```

```python
stocks = ["fb", "aapl", "nflx", "goog"]
for stock in stocks:
    print(stock.upper())
```

```python
random_things = [55, 1/2, "Puppies", stocks]
for thing in random_things:
    print(thing)
```

```python
for number in range(1, 101):
    print(f"{number} squared is {number**2}")
```

```python
squares = []
for number in range(1, 101):
    print(f"Appending {number**2} into list")
    squares.append(number**2)
print(f"The value of number is: {number}")
print(f"The value of squares is: {squares}")
```

# Dictionaries
Dictionaries are like lists but you can label each item.

```python
states = {'NY': 'New York', 'PA': 'Pennsylvania', 'CA': 'California'}
```

```python
print(states["PA"])
```

```python
user = {'name': 'Mattan', 
        'height': 70, 
        'shoe size': 10.5, 
        'hair': 'Brown', 
        'eyes': 'Brown'}
```

```python
print(f"Name: {user['name']}")
print(f"Height: {user['height']}")
print(f"Shoe Size: {user['shoe size']}")
print(f"Hair: {user['hair']}")
print(f"Eyes: {user['eyes']}")
```

```python
states["FL"] = "Florida"
print(states)
```

# Functions
Functions are just way to save code so that you can run it over and over later.

```python
def greet():
    return "Hey!"
```

```python
greet()
```

```python
def greet_by_name(name):
    return f"Hey {name}!"

def intro(person_one, person_two):
    return f"{person_one}, I'd like you to meet {person_two}"
```

```python
greet_by_name("Mattan")
```

```python
intro("Mattan", "Guido")
```
