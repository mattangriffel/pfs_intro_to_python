# Data Analysis

```python
import pandas as pd
%matplotlib inline
import seaborn as sns
import iexfinance.stocks as iex
import statsmodels.api as sm
```

```python
data = pd.read_csv('fred_data.csv', parse_dates=True, index_col=0)
data
```

```python
for column in data.columns:
    data.plot(y=column)
```

```python
sns.pairplot(data)
```

```python
sns.jointplot(x='SP500', y='DJIA', data=data, kind='reg')
```

```python
sns.heatmap(data.corr(), cmap='Blues', annot=True)
```

# Getting Stock Data from an API 

```python
xom_data = iex.get_historical_data('XOM', output_format='pandas')
```

```python
xom_data
```

# Merging Data

```python
merged_data = data.merge(xom_data, on='date')
```

```python
merged_data
```

# Linear Regression

```python
X = merged_data[['SP500', 'DGS10']].interpolate()
X = sm.add_constant(X)
y = merged_data['close']

model = sm.OLS(y, X)
results = model.fit()
results.summary()
```
